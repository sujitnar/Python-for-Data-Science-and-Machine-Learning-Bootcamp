# <span style = 'color:green'> Logistic regression part-2 </span>

```python
#we can fill in the gaps in the age column. Check for average age by passenger class.
box_age = titanic_train[['Pclass', 'Age']]
box_age.pivot(columns='Pclass', values='Age').iplot(kind='box')
```

```python
sns.boxplot(x='Pclass', y = 'Age', data = titanic_train)
#passengers in 1st and 2nd class tend to be older
```

```python
Age_1 = titanic_train.groupby('Pclass', as_index=False)['Age'].median()
Age_1 #average age of passengers by class
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass ==1:
            return 37
        if Pclass ==2:
            return 29
        if Pclass ==3:
            return 24
    else: 
        return Age
    
```

```python
#impute
titanic_train['Age'] = titanic_train[['Age', 'Pclass']].apply(impute_age, axis = 1)
```

```python
sns.heatmap(titanic_train.isnull(), yticklabels= False, cmap = 'viridis')
```

```python
#cabin has way too many missing data so drop it
titanic_train  = titanic_train.drop('Cabin', axis = 1)
```

```python
sns.heatmap(titanic_train.isnull(), yticklabels= False, cmap = 'viridis')
```

```python
titanic_train.dropna(inplace = True)
```

```python
#change the sex from male female to 0/1 and embarked from s,c,q to 0,1,2 or 1,2,3. create a dummy varialbe for each
sex = pd.get_dummies(titanic_train['Sex'], drop_first = True)
#dro_first ensures there is only a male/female column with 1 or 0 since if M = 0 is aslo F =1 and so on. 
```

```python
embark = pd.get_dummies(titanic_train['Embarked'], drop_first = True)
```

```python
embark #now Q and S are not perfect predictors since we dropped the c column
```

```python
train2 = pd.concat([titanic_train, sex], axis = 1)
train2
```

```python
train3 = pd.concat([train2,embark],axis = 1)
train3
```

```python
titanic_train = train3
titanic_train
```

```python
titanic_train.drop(['Sex', 'Name', 'Embarked', 'Ticket'], axis = 1, inplace = True)
titanic_train
```

```python
titanic_train.drop('PassengerId', axis = 1, inplace= True)
titanic_train
```

# <span style = 'color:green'> Logistic regression part-3</span>

```python
#split titanic_train into test and train for logistic regression
titanic_train.columns
```

```python
y = titanic_train['Survived']
X = titanic_train.drop('Survived', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```
